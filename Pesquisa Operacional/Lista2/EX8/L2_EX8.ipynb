```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L2_EX8', sense=LpMinimize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(2*x[1] + 2*x[2] - x[3] == 16, name='restrição_1')
model.addConstraint(3*x[1] + x[2] -x[4] == 12, name='restrição_2')

# FUNÇÃO OBJETIVA
model.setObjective(1100*x[1] + 750*x[2] )

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4]}
print(f'x = {x_sol}')
```
