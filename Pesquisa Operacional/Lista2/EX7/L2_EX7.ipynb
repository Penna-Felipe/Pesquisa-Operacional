```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L2_EX7', sense=LpMinimize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(8*x[1] + 2*x[2] - x[3] == 16, name='restrição_1')
model.addConstraint(1*x[1] + 1*x[2] - x[4] == 6, name='restrição_2')
model.addConstraint(2*x[1] + 7*x[2] - x [5] == 28, name='restrição_3')




# FUNÇÃO OBJETIVA
model.setObjective(1000*x[1] + 2000*x[2] )

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5]}
print(f'x = {x_sol}')
```
