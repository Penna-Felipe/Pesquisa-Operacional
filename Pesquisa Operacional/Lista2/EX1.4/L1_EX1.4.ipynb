```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L2_EX1.5', sense=LpMinimize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(x[1] + x[2] + x[3] == 20, name='restrição_1')
model.addConstraint(x[1] + x[2] - x[4] == 10, name='restrição_2')
model.addConstraint(5*x[1] + 6*x[2] - x[5] == 54, name='restrição_3')


# FUNÇÃO OBJETIVA
model.setObjective(10*x[1] + 12*x[2] )

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5]}
print(f'x = {x_sol}')
```
