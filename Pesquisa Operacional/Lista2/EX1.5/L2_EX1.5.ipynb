```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L2_EX1.1', sense=LpMaximize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5, 6, 7], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(-x[1] + x[2] + x[3] == 2, name='restrição_1')
model.addConstraint(x[1] + x[4] == 5, name='restrição_2')
model.addConstraint(x[2] + x[5] == 6, name='restrição_3')
model.addConstraint(3*x[1] + 5*x[2] - x[5] == 15, name='restrição_4')
model.addConstraint(5*x[1] + 4*x[2] + x[5] == 20, name='restrição_5')



# FUNÇÃO OBJETIVA
model.setObjective(7*x[1] + 9*x[2] )

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5, 6, 7]}
print(f'x = {x_sol}')
```
