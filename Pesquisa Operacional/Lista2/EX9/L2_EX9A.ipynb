```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L2_EX9A', sense=LpMaximize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(4*x[1] + 2*x[2] + x[3] == 20, name='restrição_1')
model.addConstraint(2*x[1] + 3*x[2] + x[4] == 10, name='restrição_2')
model.addConstraint(100*x[1] + 200*x[2] + x[5] == 500, name='restrição_3')

# FUNÇÃO OBJETIVA
model.setObjective(1900*x[1] + 2100*x[2] )

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5]}
print(f'x = {x_sol}')
```
