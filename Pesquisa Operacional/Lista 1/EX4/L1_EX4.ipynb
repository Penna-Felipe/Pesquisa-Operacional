```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L1_EX4', sense=LpMaximize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(x[1] + x[2] + x[3] == 5, name='restrição_1')
model.addConstraint(20*x[1] + 10*x[2] - x[4] == 80, name='restrição_2')

# FUNÇÃO OBJETIVA
model.setObjective(3000*x[1] + 10000*x[2])

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4]}
print(f'x = {x_sol}')

```
