```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L1_Ex1', sense=LpMaximize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5, 6], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(2*x[1] + x[2] + x[3] == 6, name='restrição_1')
model.addConstraint(10*x[1] + 12*x[2] + x[4] == 60, name='restrição_2')
model.addConstraint(x[1] + x[5] == 6, name='restrição_3')
model.addConstraint(x[2] + x[6] == 5, name='restrição_4')

# FUNÇÃO OBJETIVA
model.setObjective(5*x[1] + 2*x[2])

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5, 6]}
print(f'x = {x_sol}')
```
