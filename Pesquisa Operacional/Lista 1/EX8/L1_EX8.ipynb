```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L1_EX8', sense=LpMinimize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5, 6, 7], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(x[1] - x[4] == 3, name='restrição_1')
model.addConstraint(3*x[1] + 4*x[2] - x[5] == 30, name='restrição_2')
model.addConstraint(3*x[1] + 10*x[3] - x[6] == 30, name='restrição_3')
model.addConstraint(x[1] + x[2] + x[3] + x[7] == 10, name='restrição_4')

# FUNÇÃO OBJETIVA
model.setObjective(1000*x[1] + 1000*x[2] + 1000*x[3])

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5, 6]}
print(f'x = {x_sol}')
```
