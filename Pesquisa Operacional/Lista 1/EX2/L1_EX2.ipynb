```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L1_EX2', sense=LpMaximize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(2*x[1] + 3*x[2] + x[3] == 120, name='restrição_1')
model.addConstraint(x[1] + x[4] == 40, name='restrição_2')
model.addConstraint(x[2] + x[5] == 30, name='restrição_3')

# FUNÇÃO OBJETIVA
model.setObjective(100*x[1] + 150*x[2])

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5]}
print(f'x = {x_sol}')
```
