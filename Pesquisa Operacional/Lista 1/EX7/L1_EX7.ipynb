```python
#INSTALAR BIBLIOTECA PULP
!pip install pulp

from pulp import *

# DEFINIR O MODELO
model = pulp.LpProblem('L1_EX7', sense=LpMaximize)

# ADICIONAR VARIÁVEIS
x = pulp.LpVariable.dicts(indices=[1, 2, 3, 4, 5, 6], cat='LpContinous', lowBound=0, name='x')

# ADICIONAR AS RESTRIÇÕES
model.addConstraint(x[1] + x[2] + x[3] + x[4] == 100, name='restrição_1')
model.addConstraint(100*x[2] + 200*x[3] + x[5] == 14000, name='restrição_2')
model.addConstraint(100000*x[1] + 200000*x[3] + x[6] == 12750000, name='restrição_3')

# FUNÇÃO OBJETIVA
model.setObjective(300*x[1] + 400*x[2] + 500*x[3])

#OTIMIZAR
model.solve()

#OBTER E MOSTRAR RESULTADO
x_sol = {i: x[i].value() for i in [1, 2, 3, 4, 5, 6]}
print(f'x = {x_sol}')
```
